```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('/Users/martinapensalfini/Downloads/mashup_IRIS_subset_v3.xlsx')

# Step 1: Filter records where src_repo == 'amsacta'
filtered_df = df[df['src_repo'] == 'amsacta']
total_filtered = len(filtered_df)

print(f"Total records with src_repo = 'amsacta': {total_filtered}\n")

# Step 2: Count occurrences of each iris_cat value
iris_cat_counts = filtered_df['iris_cat'].value_counts()

# Step 3: Calculate percentages
iris_cat_percentages = (iris_cat_counts / total_filtered * 100).round(2)

# Step 4: Combine into a summary DataFrame
summary = pd.DataFrame({
    'count': iris_cat_counts,
    'percentage': iris_cat_percentages
})

print(summary)
```
